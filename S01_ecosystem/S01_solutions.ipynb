## Exercise 1 🟢 — Unit Converter


````python
import math

# Constants
c = 2.998e8       # m/s
h = 6.626e-34     # J·s
eV = 1.602e-19    # J/eV

# Input
lambda_nm = 532   # Nd:YAG second harmonic

# Conversions
lambda_m = lambda_nm * 1e-9
nu = c / lambda_m                         # frequency [Hz]
nu_THz = nu / 1e12                        # frequency [THz]
k = 2*math.pi / (lambda_m * 100)            # wavenumber [cm⁻¹]
E_J = h * nu                              # photon energy [J]
E_eV = E_J / eV                           # photon energy [eV]

print(f"Wavelength:    {lambda_nm:.1f} nm")
print(f"Frequency:     {nu:.3e} Hz  ({nu_THz:.1f} THz)")
print(f"Wavenumber:    {k:.0f} cm⁻¹")
print(f"Photon energy: {E_eV:.2f} eV  ({E_J:.3e} J)")
````

## Exercise 2 🟢 — Laser Spec Sheet

## Exercise 3 🟡 — Gaussian Beam Calculator


````python
import math

# Input parameters
w0_um = 50        # beam waist [μm]
lambda_nm = 800   # wavelength [nm]

# Convert to SI
w0 = w0_um * 1e-6       # [m]
lam = lambda_nm * 1e-9   # [m]

# Rayleigh range
z_R = math.pi * w0**2 / lam

# Far-field divergence (half-angle)
theta = lam / (math.pi * w0)
theta_mrad = theta * 1e3

print("=" * 55)
print(f"  Gaussian Beam Parameters")
print(f"  w₀ = {w0_um} μm, λ = {lambda_nm} nm")
print("=" * 55)
print(f"  Rayleigh range:  z_R = {z_R*1e3:.2f} mm")
print(f"  Divergence:      θ   = {theta_mrad:.3f} mrad ({math.degrees(theta):.4f}°)")
print()

# Beam size at several z positions
z_values = [1, 2, 10]  # in units of z_R
print(f"  {'z/z_R':>6s}  {'z (mm)':>10s}  {'w(z) (μm)':>10s}  {'w/w₀':>6s}")
print("  " + "-" * 38)
for z_ratio in z_values:
    z = z_ratio * z_R
    w_z = w0 * math.sqrt(1 + z_ratio**2)
    print(f"  {z_ratio:6.0f}  {z*1e3:10.2f}  {w_z*1e6:10.1f}  {w_z/w0:6.2f}")

# Also compute R(z) at z = z_R
z = z_R
R = z * (1 + (z_R / z)**2)
print(f"Radius of curvature at z = z_R: R = {R*1e3:.2f} mm")
print("=" * 55)
````

## Exercise 4 🟡 — Pulse Train Statistics


````python
import math

energies = [5.12, 4.98, 5.05, 5.21, 4.87, 5.10, 5.15, 4.93, 5.08, 5.02,
            5.18, 4.95, 5.07, 5.13, 4.89, 5.11, 5.03, 5.20, 4.96, 5.09]

N = len(energies)

# 1. Mean
mean_E = sum(energies) / N

# 2. Standard deviation
variance = sum((x - mean_E)**2 for x in energies) / N
std_E = math.sqrt(variance)

# 3. Max and min with indices
max_E = max(energies)
min_E = min(energies)
idx_max = energies.index(max_E)
idx_min = energies.index(min_E)

# 4. RMS stability
rms_stability = std_E / mean_E * 100

# 5. Report
print("=" * 45)
print(f"  {'PULSE TRAIN STATISTICS':^41}")
print("=" * 45)
print(f"  Number of pulses:  {N}")
print(f"  Mean energy:       {mean_E:.3f} nJ")
print(f"  Std deviation:     {std_E:.3f} nJ")
print(f"  Maximum:           {max_E:.3f} nJ (pulse #{idx_max + 1})")
print(f"  Minimum:           {min_E:.3f} nJ (pulse #{idx_min + 1})")
print(f"  Peak-to-peak:      {max_E - min_E:.3f} nJ")
print(f"  RMS stability:     {rms_stability:.2f}%")
print("=" * 45)

if rms_stability < 2:
    print("  → Excellent stability (< 2% RMS)")
elif rms_stability < 5:
    print("  → Good stability (< 5% RMS)")
else:
    print("  → Poor stability — check the laser!")
````

## Exercise 6 🔴 — Multi-Laser Lab Planner


````python
import math

# Constants
c = 2.998e8
h = 6.626e-34
eV_J = 1.602e-19

# Pump lasers
pumps = [
    {
        "name": "Ti:Sapph",
        "wavelength_nm": 800,
        "pulse_energy_J": 5e-3,
        "rep_rate_Hz": 1e3,
        "pulse_duration_s": 35e-15,
    },
    {
        "name": "Yb:YAG",
        "wavelength_nm": 1030,
        "pulse_energy_J": 200e-6,
        "rep_rate_Hz": 100e3,
        "pulse_duration_s": 300e-15,
    },
    {
        "name": "Nd:YAG",
        "wavelength_nm": 1064,
        "pulse_energy_J": 100e-3,
        "rep_rate_Hz": 10,
        "pulse_duration_s": 5e-9,
    },
]

# Probe lasers
probes = [
    {"name": "HeNe", "wavelength_nm": 632.8, "type": "CW", "power_W": 5e-3},
    {"name": "Diode", "wavelength_nm": 405, "type": "CW", "power_W": 50e-3},
    {"name": "White light", "wavelength_nm": 650, "type": "SC", "power_W": 1.0},  # center wavelength
]

# Focus parameters
w0_m = 20e-6  # 20 μm focus

# 2. Compute peak power and peak intensity for pumps
for p in pumps:
    p["peak_power_W"] = 0.94 * p["pulse_energy_J"] / p["pulse_duration_s"]
    p["peak_intensity_Wcm2"] = p["peak_power_W"] / (math.pi * (w0_m * 100)**2)  # w0 in cm
    p["photon_energy_eV"] = h * c / (p["wavelength_nm"] * 1e-9) / eV_J

for p in probes:
    p["photon_energy_eV"] = h * c / (p["wavelength_nm"] * 1e-9) / eV_J

# 3. Rank by peak intensity
pumps_ranked = sorted(pumps, key=lambda x: x["peak_intensity_Wcm2"], reverse=True)

print("=" * 70)
print(f"  {'PUMP LASERS — Ranked by Peak Intensity at Focus':^66}")
print(f"  (w₀ = {w0_m*1e6:.0f} μm)")
print("=" * 70)
print(f"  {'Name':>10s}  {'P_peak (W)':>12s}  {'I_peak (W/cm²)':>16s}  {'E_photon (eV)':>14s}")
print("  " + "-" * 58)
for p in pumps_ranked:
    print(f"  {p['name']:>10s}  {p['peak_power_W']:12.3e}  {p['peak_intensity_Wcm2']:16.3e}  {p['photon_energy_eV']:14.3f}")

# 4 & 5. Pump-probe energy differences
print(f"{'=' * 55}")
print(f"  {'PUMP-PROBE PHOTON ENERGY DIFFERENCES':^51}")
print("=" * 55)
print(f"  {'Pump':>10s}  {'Probe':>12s}  {'ΔE (eV)':>10s}")
print("  " + "-" * 36)

best_combo = None
min_dE = float("inf")

for pump in pumps:
    for probe in probes:
        dE = abs(pump["photon_energy_eV"] - probe["photon_energy_eV"])
        print(f"  {pump['name']:>10s}  {probe['name']:>12s}  {dE:10.3f}")
        if dE < min_dE:
            min_dE = dE
            best_combo = (pump["name"], probe["name"])

print(f"  → Closest match: {best_combo[0]} + {best_combo[1]} (ΔE = {min_dE:.3f} eV)")
print("=" * 55)
````

## Exercise 7 🔴 — Dispersion Calculator


````python
import math

# BK7 parameters at 800 nm
GVD = 44.7       # fs²/mm
tau0 = 30         # initial FWHM [fs]

thicknesses = [1, 2, 3, 4, 5, 10, 15, 20, 30, 50]

print("=" * 50)
print(f"  Pulse Broadening in BK7 Glass")
print(f"  τ₀ = {tau0} fs, GVD = {GVD} fs²/mm")
print("=" * 50)
print(f"  {'L (mm)':>8s}  {'τ(L) (fs)':>10s}  {'τ/τ₀':>8s}  {'Broadening':>12s}")
print("  " + "-" * 42)

results = []
critical_L = None

for L in thicknesses:
    # τ(L) = τ₀ * sqrt(1 + (4 ln2 · GVD · L / τ₀²)²)
    ratio = 4 * math.log(2) * GVD * L / tau0**2
    tau_L = tau0 * math.sqrt(1 + ratio**2)
    broadening = (tau_L / tau0 - 1) * 100
    results.append((L, tau_L))
    
    marker = " ◄ 10%" if abs(broadening - 10) < 2 else ""
    print(f"  {L:8d}  {tau_L:10.2f}  {tau_L/tau0:8.3f}  {broadening:10.1f}%{marker}")

# Find critical thickness (broadening < 10%)
# Solve: sqrt(1 + x²) = 1.1 → x² = 0.21 → x = 0.458
# x = 4 ln2 · GVD · L / τ₀²
# L = x · τ₀² / (4 ln2 · GVD)
x_crit = math.sqrt(1.1**2 - 1)
L_crit = x_crit * tau0**2 / (4 * math.log(2) * GVD)

print(f"  Critical thickness (< 10% broadening): {L_crit:.1f} mm")
print(f"  Rounded down: {math.floor(L_crit)} mm")
print("=" * 50)
````

> 🔬 **Physics Insight:** For a 30 fs pulse at 800 nm, even a few mm of glass causes significant broadening. This is why femtosecond laser systems use chirped mirrors and prism compressors to manage dispersion — a topic you'll simulate in later sessions using NumPy and SciPy.
