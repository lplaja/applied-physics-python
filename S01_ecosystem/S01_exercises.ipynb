## Exercise 1 🟢 — Unit Converter

Electromagnetic radiation can be described by its wavelength (λ), frequency (ν), wavenumber (k), or photon energy (E). Write code that, given a wavelength in nm, prints all four representations.

**Physical relations:**
$$\nu = \frac{c}{\lambda}, \quad k = \frac{2 \pi}{\lambda}, \quad E = h\nu$$

**Constants:**
- $c = 2.998 \times 10^8$ m/s
- $h = 6.626 \times 10^{-34}$ J·s
- $1\,\text{eV} = 1.602 \times 10^{-19}$ J

**Test with:** λ = 532 nm (Nd:YAG second harmonic)

**Expected output format:**
```
Wavelength:    532.0 nm
Frequency:     5.635e+14 Hz
Wavenumber:    118105 cm⁻¹
Photon energy: 2.33 eV
```


````python
# Exercise 1 — your code here

import math

c = 2.998e8       # m/s
h = 6.626e-34     # J·s
eV = 1.602e-19    # J/eV

lambda_nm = 532   # input wavelength

# Convert and print all representations:

lamb=lambda_nm*1e-9
nu=c/lamb
k=2*math.pi/lambda_nm*1/1e-7
E=h*nu

print(f'Wavelength \t {lambda_nm:.1f} ')
print(f'Frequency \t {nu:.3e} Hz ')
print(f'Wavenumber \t {k:.0f} cm⁻¹ ')
print(f'Photon energy \t {E/eV:.2f} eV ')
````

## Exercise 7 🔴 — Dispersion Calculator

When an ultrashort pulse propagates through a dispersive material, it acquires a spectral phase. The key parameters are the Group Velocity Dispersion (GVD, $\phi''$) and Third-Order Dispersion (TOD, $\phi'''$).

For **BK7 glass** at 800 nm:
- GVD: $\phi'' = 44.7\,\text{fs}^2/\text{mm}$
