## Exercise 3 🟢 — Fresnel equations

The Fresnel equations give reflectance at an interface between two media (refractive indices `n1`, `n2`) at angle of incidence `θ_i` (radians):

$$R_s = \left(\frac{n_1 \cos\theta_i - n_2 \cos\theta_t}{n_1 \cos\theta_i + n_2 \cos\theta_t}\right)^2, \quad
R_p = \left(\frac{n_2 \cos\theta_i - n_1 \cos\theta_t}{n_2 \cos\theta_i + n_1 \cos\theta_t}\right)^2$$

where $\theta_t$ is given by Snell's law: $n_1 \sin\theta_i = n_2 \sin\theta_t$.

## Exercise 9 🔴 — Debugging challenge

The function below is supposed to compute the **transform-limited pulse duration** from a given spectral bandwidth using the time-bandwidth product for a Gaussian pulse: $\Delta t \cdot \Delta\nu = 0.4413$.

It has **three bugs**. Use the VS Code debugger (not print statements) to find and fix all three. Document each bug and your fix in a comment.

Expected results:
- Spectral bandwidth 10 nm at 800 nm → ~94 fs transform limit
- Spectral bandwidth 5 nm at 1030 nm → ~283 fs transform limit

````python
import math

def transform_limit(delta_lambda_nm, center_lambda_nm):
    """
    Transform-limited pulse duration for a Gaussian pulse.

    Uses: Δt · Δν = 0.4413  (Gaussian pulse time-bandwidth product)
    and:  Δν ≈ (c / λ²) · Δλ

    Parameters
    ----------
    delta_lambda_nm : float
        Spectral bandwidth FWHM [nm].
    center_lambda_nm : float
        Center wavelength [nm].

    Returns
    -------
    float
        Transform-limited pulse duration FWHM [fs].
    """
    TBP = 0.4413
    c   = 3e8   # m/s

    # Convert to SI
    dl = delta_lambda_nm * 1e-9   # m
    lc = center_lambda_nm         # BUG 1: forgot unit conversion

    # Frequency bandwidth: Δν ≈ (c/λ²) · Δλ
    d_nu = (c / lc) * dl          # BUG 2: wrong formula

    # Transform limit
    dt_s = TBP * d_nu             # BUG 3: wrong relationship between TBP, Δt, Δν

    return dt_s * 1e15   # convert to fs


# Tests
print(f"10 nm @ 800 nm  → {transform_limit(10, 800):.1f} fs  (expected ~94 fs)")
print(f" 5 nm @ 1030 nm → {transform_limit(5, 1030):.1f} fs  (expected ~283 fs)")
````
